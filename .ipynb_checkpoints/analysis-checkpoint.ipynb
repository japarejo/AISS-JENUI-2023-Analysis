```python
import pandas as pd
import plot_likert as pl
import matplotlib.pyplot as plt
df2021 = pd.read_excel("data/AISS-2021-Encuesta-arquitectura-anonimizada.xlsx")
df2022 = pd.read_excel("data/AISS-2022-Encuesta-arquitectura-anonimizada.xlsx")
# We fix some minor typos in the survey of 2022
df2022.replace([' Muy insatisfecho/a'], 'Muy insatisfecho/a')
df=pd.concat([df2021,df2022]);
# We Homogeneize column names:



# We fix some minor typos in the survey of 2022
print(df2021.columns)
print(df2022.columns)
```

```python
myscale = \
    ['Muy insuficiente',
     'Insuficiente',
     'Suficiente',
     'Adecuada',
     'Muy adecuada']

myplot2021=pl.plot_likert(df2021[df2021.columns[5:13]], myscale, plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot2021.xaxis.set_label_text('Porcentaje de respuestas en 2021')
myplot2021.get_figure().savefig('figs/SatisfaccionFormacionPorApartado2021.png')
myplot2021
```

```python
myplot2022=pl.plot_likert(df2022[df2022.columns[5:13]], myscale, plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot2022.xaxis.set_label_text('Porcentaje de respuestas en 2022')
myplot2022.get_figure().savefig('figs/SatisfaccionFormacionPorApartado2022.png');
myplot2022
```

```python
myplot=pl.plot_likert(df[df.columns[5:13]], myscale, plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot.xaxis.set_label_text('Porcentaje de respuestas en ambos cursos')
myplot.get_figure().savefig('figs/SatisfaccionFormacionPorApartado.png');
myplot
```

```python
# Niveles de satisfacción en escala de likert:
satisfaction_scale = \
    ['Muy insatisfecho/a',
     'Algo insatisfecho/a',
     'Ni satisfecho/a ni insatisfecho/a',
     'Algo satisfecho/a',
     'Muy satisfecho/a']
df.rename(columns = {'Columna1':'Satisfacción general',
                        'Columna2':'Claridad de los criterios de evaluación del proyecto',
                        'Columna3':'Nivel de detalle de los criterios de evaluación del proyecto',
                        'Columna4':'Plantilla del proyecto',
                        'Columna5':'Proyecto de referencia',
                        'Columna6':'Nivel de complejidad del proyecto asignado'},
          inplace = True)
myplot=pl.plot_likert(df[[
                          'Claridad de los criterios de evaluación del proyecto',
                          'Nivel de detalle de los criterios de evaluación del proyecto',
                          'Plantilla del proyecto',
                          'Proyecto de referencia',
                          'Nivel de complejidad del proyecto asignado',
                          'Satisfacción general'
                         ]], 
                      satisfaction_scale, 
                      plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot.xaxis.set_label_text('Porcentaje de satisfacción en ambos cursos')
myplot.get_figure().savefig('figs/SatisfaccionGeneralLikert.png');
myplot
```

```python
# Niveles de satisfacción en escala de likert:
utility_scale = \
    ['Completamente inútil',
     'Algo inútil',
     'Ni útil ni inútil',
     'Algo útil',
     'Muy útil']
df.rename(columns = {'Columna9':'Las métricas analizadas',
                     'Columna10':'La propia herramienta SonarQube'
                     },
          inplace = True)
myplot=pl.plot_likert(df2021[                          
                          df2021.columns[[35,36,37,38]]
                         ], 
                      utility_scale, 
                      plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot.xaxis.set_label_text('Utilidad percibida en ambos cursos')
myplot.get_figure().savefig('figs/SatisfaccionGeneralLikert.png');
myplot
```

```python
values=df2021['Satisfacción general con la formación'].value_counts()
values
plt.title('Nivel de satisfacción con general con la formación en 2021')
myplot=plt.pie(values,labels=values.index.tolist(),autopct= '%1.1f%%')
plt.savefig('figs/SatisfaccionGeneral2021.png')
myplot
```
