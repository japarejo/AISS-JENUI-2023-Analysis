## Data import 

Prior to any other data analysis, the files have been anonymized, removing any column that could provide information about the authors of the responses, and substituting them by an integer identifier.
The data of the surveys is provided  in two different files: 
 * AISS-2021-Encuesta-arquitectura-anonimizada: Which contains  the data corresponding to the course 2020-2021
 * AISS-2022-Encuesta-arquitectura-anonimizada: Which contains  the data corresponding to the course 2021-2022
The files are provided in excel format  (with extension .xlsx), thus we use pandas to import  them. 
Additionally, we use  the packages 'plot_likert' and 'matplotlib' to generate charts.

```python
import pandas as pd
import plot_likert as pl
import matplotlib.pyplot as plt


#===================#
#    DATA IMPORT    #
#===================#
df2021 = pd.read_excel("data/AISS-2021-Encuesta-arquitectura-anonimizada.xlsx")
df2022 = pd.read_excel("data/AISS-2022-Encuesta-arquitectura-anonimizada.xlsx")
display(df2021)
display(df2022)
```

## Data Wrangling
Next we modify the structure of the data-frames, since office 365 introduces some spureous columns, and some of the columns names are not appropirate for data analysis and reporting.
Additionally, we fix some minor typos present in the response values of the second survey (2022).

```python
#=======================#
#    DATA WRANGLING     #
#=======================#
#------------------------#
#    COLUMN REMOVAL      #
#------------------------#
columnsToRemove2021=['Pregunta','Pregunta2', 'Pregunta3', 'X','Instrucción 2', 'Pregunta4','Instrucción 22',
                     'Instrucción 23','Instrucción 24','Pregunta5','¿Qué probabilidades hay de que nos recomiende a un amigo o compañero?']
columnsToRemove2022=['Vista de despliegue2','Vista funcional2','Vista de desarrollo2','Vista de contexto2',
                     'Puntos de variabilidad y extensión2','Vista de escenarios de uso2','Presentación2',
                     'Análisis de atributos de calidad2','Sugerencias de mejora y contribuciones2','X2']
df2021=df2021.drop(columns=columnsToRemove2021,errors='ignore')
df2022=df2022.drop(columns=columnsToRemove2022,errors='ignore')
#------------------------#
#    COLUMN RENAMIG      #
#------------------------#
columnsToRename2021={'Columna1':'Satisfacción general con la formación',
                     'Columna2':'Claridad de los criterios de evaluación',
                     'Columna3':'Nivel de detalle de los criterios de evaluación',
                     'Columna4':'Plantilla del proyecto',
                     'Columna5':'Proyecto de referencia',
                     'Columna6':'Nivel de complejidad del proyecto asignado',
                     'Columna7':'Conocimientos de Github antes',
                     'Columna8':'Conocimientos de Github después',
                     'Columna9':'Utilidad de métricas analizadas',
                     'Columna10':'Utilidad de SonarQube',
                     'Columna11':'Utilidad de presentar los proyectos',
                     'Columna12':'Satisfacción con la metodología empleada',
                     'Vista de despliegue':'Utilidad de la vista de despliegue',
                     'Vista de contexto':'Utilidad de la vista de contexto',
                     'Vista funcional':'Utilidad de la vista funcional',
                     'Vista de escenarios de uso':'Utilidad de la vista de escenarios de uso',
                     'Vista de despliegue ':'Vista de despliegue',
                     'Vista de contexto ':'Vista de contexto',
                     'Vista de funcional ':'Vista funcional',
                     'Vista de desarrollo ':'Vista de desarrollo',
                     'Vista de desarrollo 2':'Utilidad de la vista de desarrollo',
                     'Análisis de atributos de calidad2':'Utilidad de la analizar los atributos de calidad',
                     'Vista de escenarios de uso ':'Vista de escenarios de uso',
                     'Análisis de atributos de calidad':'Utilidad de la análizar los atributos de calidad',
                     'Puntos de variabilidad y extensión ':'Utilidad de los puntos de variabilidad y extensión',
                     'Utilidad de métricas analizadas':'Utilidad de las métricas analizadas'}

columnsToRename2022={'Columna1':'Satisfacción general con la formación',
                     'Columna3':'Claridad de los criterios de evaluación',
                     'Columna2':'Nivel de detalle de los criterios de evaluación',
                     'Columna4':'Plantilla del proyecto',
                     'Columna5':'Proyecto de referencia',
                     'Columna6':'Nivel de complejidad del proyecto asignado',
                     'Columna7':'Conocimientos de Github antes',
                     'Columna8':'Utilidad de métricas analizadas',
                     'Columna9':'Conocimientos de Github después',
                     'Columna10':'Utilidad de presentar los proyectos',
                     'Columna11':'Utilidad de SonarQube',
                     'Columna12':'Satisfacción con la metodología empleada',
                     'Vista de contexto3':'Utilidad de la vista de contexto',
                     'Vista de escenarios de uso\n':'Utilidad de la vista de escenarios de uso',
                     'Vista funcional\n':'Utilidad de la vista funcional',
                     'Vista de despliegue\n':'Utilidad de la vista de despliegue',
                     'Vista de desarrollo\n':'Utilidad de la vista de desarrollo',
                     'Puntos de variabilidad y extensión\n':'Utilidad de los puntos de variabilidad y extensión',
                     'Análisis de atributos de calidad\n':'Utilidad de la analizar los atributos de calidad',
                     'Utilidad de métricas analizadas':'Utilidad de las métricas analizadas'}
df2021=df2021.rename(columns=columnsToRename2021)
df2022=df2022.rename(columns=columnsToRename2022)
print(df2021.columns)
print(df2022.columns)
#------------------------#
#    COLUMN SWAPPING     #
#------------------------#
columnsToSwap2021={}
columnsToSwap2022={}#'Vista de contexto':'Participantes'}

#define function to swap columns
def swap_columns(df, col1, col2):
    col_list = list(df.columns)
    x, y = col_list.index(col1), col_list.index(col2)
    col_list[y], col_list[x] = col_list[x], col_list[y]
    df = df[col_list]
    return df

for x in columnsToSwap2021:
    print(x)
    print(columnsToSwap2021[x])
    df2021=swap_columns(df2021,x,columnsToSwap2021[x])
    
for x in columnsToSwap2022:
    print(x)
    print(columnsToSwap2022[x])
    df2022=swap_columns(df2022,x,columnsToSwap2022[x])

#------------------------#
#   Generate Unique IDs  #
#------------------------#
df2021['ID']=df2021['ID']+2021000
df2022['ID']=df2022['ID']+2022000
#------------------------#
#    TYPOS FIXING        #
#------------------------#
# We fix some minor typos in the survey of 2022
df2022=df2022.replace([' Muy insatisfecho/a'], 'Muy insatisfecho/a')
df=pd.concat([df2021,df2022]);
df
```

```python
#================================#
#    LIKERT TEACHING ANALYSIS    #
#================================#

myscale = \
    ['Muy insuficiente',
     'Insuficiente',
     'Suficiente',
     'Adecuada',
     'Muy adecuada']

myplot2021=pl.plot_likert(df2021[df2021.columns[4:13]], myscale, plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot2021.xaxis.set_label_text('Porcentaje de respuestas en 2021')
myplot2021.get_figure().savefig('figs/SatisfaccionFormacionPorApartado2021.png',bbox_inches='tight',dpi=300)
myplot2021
```

```python
myplot2022=pl.plot_likert(df2022[df2022.columns[4:13]], myscale, plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot2022.xaxis.set_label_text('Porcentaje de respuestas en 2022')
myplot2022.get_figure().savefig('figs/SatisfaccionFormacionPorApartado2022.png',bbox_inches='tight',dpi=300);
myplot2022
```

```python
myplot=pl.plot_likert(df[df.columns[4:13]], myscale, plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot.xaxis.set_label_text('Satisfacción con la formación recibida por apartados en ambos cursos')
myplot.get_figure().savefig('figs/SatisfaccionConLaFormacionPorApartado.png',bbox_inches='tight',dpi=300);
myplot
```

```python
# Niveles de satisfacción en escala de likert:
satisfaction_scale = \
    ['Muy insatisfecho/a',
     'Algo insatisfecho/a',
     'Ni satisfecho/a ni insatisfecho/a',
     'Algo satisfecho/a',
     'Muy satisfecho/a']
df.replace(df['Claridad de los criterios de evaluación'].unique()[3],'Muy insatisfecho/a',inplace=True)

myplot=pl.plot_likert(df[[                          
                         'Claridad de los criterios de evaluación',
                          'Nivel de detalle de los criterios de evaluación',
                          'Plantilla del proyecto',
                          'Proyecto de referencia',
                          'Nivel de complejidad del proyecto asignado',
                          'Satisfacción general con la formación'
                         ]], 
                      satisfaction_scale, 
                      plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot.xaxis.set_label_text('Satisfacción con los elementos del proyecto en ambos cursos')
myplot.get_figure().savefig('figs/SatisfaccionGeneralLikert.png',bbox_inches='tight',dpi=300);
myplot
```

```python
# Utility levels in likert scale:
utility_scale = \
    ['Completamente inútil',
     'Algo inútil',
     'Ni útil ni inútil',
     'Algo útil',
     'Muy útil']
df=df.replace('Algo útiles','Algo útil')
df=df.replace('Algo útil\n','Algo útil')
df=df.replace('Algo inútiles','Algo inútil',)
df=df.replace('Algo inútil\n','Algo inútil',)
df=df.replace('Ni útiles ni inútiles','Ni útil ni inútil')
df=df.replace('Muy útiles','Muy útil')
df=df.replace('Muy útil\n','Muy útil')
df=df.replace('Completamente inútiles','Completamente inútil')
df=df.replace('Completamente inútil\n','Completamente inútil')
myplot=pl.plot_likert(df[[                          
                            'Utilidad de la vista de contexto',
                            'Utilidad de la vista de escenarios de uso',
                            'Utilidad de la vista funcional',
                            'Utilidad de la vista de desarrollo',
                            'Utilidad de la vista de despliegue',
                            'Utilidad de los puntos de variabilidad y extensión',                                                        
                            'Utilidad de métricas analizadas',
                            'Utilidad de la analizar los atributos de calidad',
                            'Utilidad de SonarQube',
                            'Utilidad de presentar los proyectos'
                         ]], 
                      utility_scale, 
                      plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot.set_xlim([-10,90])
extraticks=[0,-10,-20]
#print(myplot.get_xticklabels())
myplot=myplot.set_xticks(list(myplot.get_xticks()) + extraticks)
#myplot=myplot.set_xticklabels(list(myplot.get_xticklabels())+["30%","40%","50%"])
myplot.xaxis.set_label_text('Utilidad percibida en ambos cursos')
myplot.get_figure().savefig('figs/UtilidadPorElementosDelProyectoLikert.png',bbox_inches='tight',dpi=300);
myplot
```

```python
replacer = {'Muy insuficiente': 1, 'Insuficiente': 2,
            'Suficiente': 3,
            'Adecuado': 4,'Adecuado ': 4, 'Avanzado': 5}

df["github-antes"]=df["Conocimientos de Github antes"].replace(replacer)
df["github-despues"]=df["Conocimientos de Github después"].replace(replacer)
df["diferencia-github"]=df["github-despues"]-df["github-antes"]
values=df['diferencia-github'].value_counts()
print(values)
plt.title('Mejora  en niveles de likert  generada en cuanto a los conocimientos de Github')
plt.pie(values,labels=values.index.tolist(),autopct= '%1.1f%%')
```
