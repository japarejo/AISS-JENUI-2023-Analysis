```python
import pandas as pd
import plot_likert as pl
import matplotlib.pyplot as plt
df2021 = pd.read_excel("data/AISS-2021-Encuesta-arquitectura-anonimizada.xlsx")
df2022 = pd.read_excel("data/AISS-2022-Encuesta-arquitectura-anonimizada.xlsx")
df=pd.concat([df2021,df2022]);
```

```python
myscale = \
    ['Muy insuficiente',
     'Insuficiente',
     'Suficiente',
     'Adecuada',
     'Muy adecuada']

myplot2021=pl.plot_likert(df2021[df2021.columns[5:13]], myscale, plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot2021.xaxis.set_label_text('Porcentaje de respuestas en 2021')
myplot2021.get_figure().savefig('figs/SatisfaccionFormacionPorApartado2021.png');
myplot2021
```

```python
myplot2022=pl.plot_likert(df2022[df2022.columns[5:13]], myscale, plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot2022.xaxis.set_label_text('Porcentaje de respuestas en 2022')
myplot2022.get_figure().savefig('figs/SatisfaccionFormacionPorApartado2022.png');
myplot2022
```

```python
myplot=pl.plot_likert(df[df.columns[5:13]], myscale, plot_percentage=True, bar_labels=True, bar_labels_color="snow")
myplot.xaxis.set_label_text('Porcentaje de respuestas en ambos cursos')
myplot.get_figure().savefig('figs/SatisfaccionFormacionPorApartado.png');
myplot
```
